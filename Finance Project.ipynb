```python
start = datetime.datetime(2006, 1, 1)
end = datetime.datetime(2016, 1, 1)
```

```python
BAC = data.DataReader("BAC", 'stooq', start, end)
# CitiBank
C = data.DataReader("C", 'stooq', start, end)
# Goldman Sachs
GS = data.DataReader("GS", 'stooq', start, end)
# JP Morgan Chase
JPM = data.DataReader("JPM", 'stooq', start, end)
# Morgan Stanley
MS = data.DataReader("MS", 'stooq', start, end)
# Wells Fargo
WFC = data.DataReader("WFC", 'stooq', start, end)
```

```python
df = data.DataReader(['BAC', 'C', 'GS', 'JPM', 'MS', 'WFC'], 'stooq', start, end)
```

```python
# df.columns
```

** Creating a list of the ticker symbols (as strings) in alphabetical order. Call this list: tickers**

```python
tickers = ['BAC', 'C', 'GS', 'JPM', 'MS', 'WFC']
```

```python
df
```

** Using pd.concat to concatenate the bank dataframes together to a single data frame called bank_stocks. Setting the keys argument equal to the tickers list.**

```python
bank_stocks = pd.concat([BAC, C, GS, JPM, MS, WFC], axis =1, keys=tickers)
bank_stocks
```

```python
bank_stocks.columns.names = ['Bank Ticker','Stock Info']
```

```python
bank_stocks.head()
```

# EDA

Let's explore the data a bit!

** Checking the max Close price for each bank's stock throughout the time period**

```python
bank_stocks.xs(key='Close', axis=1, level='Stock Info').max()
```

```python
dictionary = {-0.02300886942418079: 8.76991150442478,
 0.008928605640421639: -0.8991071428571429,
 0.09695019542210614: 0.08410914927768864,
 -0.0002926052697974635: 0.011548711874444617,
 0.14322252868437624: 0.14833759590792828}
```

```python
returns = pd.DataFrame()
```

** Creating a new empty DataFrame called returns. This dataframe will contain the returns for each bank's stock. returns are typically defined by:**

$$r_t = \frac{p_t - p_{t-1}}{p_{t-1}} = \frac{p_t}{p_{t-1}} - 1$$

** Using pandas pct_change() method on the Close column to create a column representing this return value. Creating a for loop that goes and for each Bank Stock Ticker creates this returns column and set's it as a column in the returns DataFrame.**

```python
for tick in tickers:
    returns[tick+' Return'] = bank_stocks[tick]['Close'].pct_change()
returns.head()
```

```python
returns['C Return'].replace(dictionary,inplace=True)
```

```python
import matplotlib.pyplot as plt
for tick in tickers:
    bank_stocks[tick]['Close'].plot(figsize=(12, 4), label=tick)
plt.legend()
```

## Using Plotly and Cufflinks

In this part of the project there is use of cufflinks library to create some Technical Analysis plots. This part of the project is experimental due to its heavy reliance on the cuffinks project.

** Using .iplot(kind='candle) to create a candle plot of Bank of America's stock from Jan 1st 2015 to Jan 1st 2016.**

```python
BAC[['Open', 'High', 'Low', 'Close']].sort_index().loc['2015-01-01':'2016-01-01'].iplot(kind='candle')
```

** Using .ta_plot(study='sma') to create a Simple Moving Averages plot of Morgan Stanley for the year 2015.**

```python
MS['Close'].sort_index().loc['2015-01-01':'2016-01-01'].ta_plot(study='sma',periods=[13,21,55],title='Simple Moving Averages')
```

**Using .ta_plot(study='boll') to create a Bollinger Band Plot for Bank of America for the year 2015.**

```python
BAC['Close'].sort_index().loc['2015-01-01':'2016-01-01'].ta_plot(study='boll')
```
